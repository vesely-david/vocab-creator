# EbookLib

```python
import ebooklib
from ebooklib import epub
```

```python
book = epub.read_epub("/Users/david.vesely/Dropbox/My books/zero.to.one.notes.on.startups.or.how.to.build.the.future.epub")
```

```python
book.get_metadata('DC', '')
```

```python
items = list(book.get_items_of_type(ebooklib.ITEM_DOCUMENT))
```

```python
len(items)
```

```python
chapters = [item for item in items if item.is_chapter()]
```

```python
len(chapters)
```

```python
[ch.title for ch in chapters if len(ch.title) > 0]
```

```python
navigations = list(book.get_items_of_type(ebooklib.ITEM_NAVIGATION))
navigations
```

```python
for item in book.get_items():
    if item.get_type() == ebooklib.ITEM_DOCUMENT:
        print('==================================')
        print('NAME : ', item.get_name())
        print('----------------------------------')
        print(item.get_content())
        print('==================================')
```
